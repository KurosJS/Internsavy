## Exploratory Data Analysis

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data
df = pd.read_csv('./data/Admission_Predict_Ver2.csv')
df.head()
```

```python
#drop the serial no. column as it is not a meaningful feature
df.drop('Serial No.', axis=1, inplace=True)
```

```python
#plotting histograms to see the distribution of the data and to check for outliers
df.hist(bins=25, figsize=(20,20), color='r', edgecolor='black', alpha = 0.5)
plt.show()
```

### Quick overview of the Histograms
We can see that the data is mostly distributed without too extreme values, however in the LOR (Letter of Recommendation strength) and CGPA (Undergraduate GPA) low values can be observed.

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 20))

# Create box plots for the GRE score
axs[0].boxplot(df['GRE Score'])
axs[0].set_xticklabels(['GRE Score'])
axs[0].set_title('GRE Score')

# Create box plots for the TOEFL score
axs[1].boxplot(df['TOEFL Score'])
axs[1].set_xticklabels(['TOEFL Score'])
axs[1].set_title('TOEFL Score')

# Create box plots for the remaining columns
axs[2].boxplot([df['University Rating'], df['SOP'], df['LOR'], df['CGPA'], df['Research']])
axs[2].set_xticklabels(['University Rating', 'SOP', 'LOR', 'CGPA', 'Research'])
axs[2].set_title('Other data')

# Display the plot
plt.show()
```

### Quick overview of the Boxplots
The boxplots show us that SOP (Statement of Purpose strength) has a great difference between minimum value and median, which means that there are some outliers in the data. The same can be said about LOR and CGPA, but to a lesser extent. Research values are binary, so there is no need to check for outliers. Data from other columns is distributed without outliers.
